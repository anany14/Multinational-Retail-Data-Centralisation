```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

factorial(6)
```

```python
def fibonacci(n):
    if n == 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        fib_list = fibonacci(n-1)
        fib_list.append(fib_list[-1] + fib_list[-2])
        return fib_list

fibonacci(6)
```

```python
import math

class Cylinder:
    def __init__(self, height, radius=1):
        self.height = height
        self.radius = radius
        self.surface_area = self.get_surface_area()
        self.volume = self.get_volume()

    def get_surface_area(self):
        sa = 2 * math.pi * self.radius * (self.radius + self.height)
        return round(sa, 2)

    def get_volume(self):
        vol = math.pi * self.radius**2 * self.height
        return round(vol, 2)

bruh = Cylinder(5,1)
print(bruh.surface_area)
```

```python
class Phone:
    def __init__(self,number,balance=20):
        self.number = number
        self.balance = balance
        

    def make_call(self,number,duration=1):

        if self.balance >= duration:
            print(f"calling {number} for {duration} minutes")
            self.balance -= duration
        else:
            print("insufficient balance")

    def info(self):
        print(self.number,self.balance)

call = Phone(111,2)
call.info()
call.make_call(222,1)





```

```python
import pandas as pd
from sqlalchemy import create_engine, Column, Integer, String
from sqlalchemy.ext.declarative import declarative_base

#already made the engine

# Define a declarative base class to use for defining the tables
Base = declarative_base()

# Define a class to represent the Samuel L. Jackson movies table
class SamuelMovies(Base):
    __tablename__ = 'samuel_movies'
    title = Column(String, primary_key=True)
    year = Column(Integer)

# Define a class to represent the Tim Roth movies table
class TimMovies(Base):
    __tablename__ = 'tim_movies'
    title = Column(String, primary_key=True)
    year = Column(Integer)

# Load CSV data into dataframes
jackson_df = pd.read_csv("samuel_movies.csv")
roth_df = pd.read_csv("tim_movies.csv")

# Write dataframes to the database
jackson_df.to_sql('samuel_movies', engine, if_exists='replace', index=False)
roth_df.to_sql('tim_movies', engine, if_exists='replace', index=False)

# Commit the changes and close the session
engine.dispose()
```
